```python
import numpy as np
from scipy import signal as sg
import matplotlib.pyplot as plt
import matplotlib as mpl
label_size = 15
mpl.rcParams['xtick.labelsize'] = label_size 
mpl.rcParams['ytick.labelsize'] = label_size 
```

```python
# signal sampling frequency
sf = 500
# signal time length
tl = [0, 1]
# frequency signal 1
f1 = 10
# frequency signal 2
f2 = 5
```

```python
# time vector
t = np.linspace(tl[0], tl[1], sf)
# signal 1
s1 = np.sin(2 * np.pi * f1 * t)
# signal 2
s2 = np.cos(2 * np.pi * f2 * t)
```

```python
plt.figure(figsize=[8,8])
plt.subplot(2,1,1)
plt.title('S1 = 10Hz Signal', fontsize=20)
plt.plot(t, s1)
plt.ylabel('Amplitude', fontsize=20)
plt.subplot(2,1,2)
plt.title('S2 = 5Hz Signal', fontsize=20)
plt.plot(t,s2)
plt.ylabel('Amplitude', fontsize=20)
plt.xlabel('Time [sec]', fontsize=20)

plt.show()
```

```python
# add the two signals
plt.figure(figsize=[8,8])
plt.title('S1 + S2', fontsize=20)
s3 = s1 + s2
plt.plot(t, s3)
plt.ylabel('Amplitude', fontsize=20)
plt.xlabel('Time [sec]', fontsize=20)
# plt.xticks(fontsize=20)
plt.show()
```

```python
plt.figure(figsize=[8,8])
plt.title('Spectrogram Signal S1+S2', fontsize=20)
f, t, Sxx = sg.spectrogram(s3, 500)
plt.pcolormesh(t, f, Sxx, shading='gouraud')
plt.ylabel('Frequency [Hz]', fontsize=20)
plt.xlabel('Time [sec]', fontsize=20)
plt.ylim([0, 20])
plt.show()
```

This signal from zero to one has a constant combined frequency of 5Hz and 10Hz. If these frequency components change after the first second, would the spectrogram capture these changes?

Lets create another signal with a different frequency components and then add it to the end of our signal. This will represent a signal which frequency components change in time.

```python
# frequency of the signal
f4 = 20
# signal
s4 = np.sin(2 * np.pi * f4 * t)
# add it in the end of the first signal
s5 = np.hstack([s3, s4])
```

```python
len(s5)
```

```python
plt.figure(figsize=[10,10])
plt.subplot(2,1,1)
plt.title('Time-Amplitud Signal', fontsize=20)
plt.plot(np.linspace(0, 2, 2*sf), s5)
plt.ylabel('Amplitud', fontsize=20)

plt.subplot(2,1,2)
plt.title('Signal Spectrogram', fontsize=20)
f, t, Sxx = sg.spectrogram(s5, 500)
plt.pcolormesh(t, f, Sxx, shading='gouraud')
plt.ylabel('Frequency [Hz]', fontsize=20)
plt.xlabel('Time [sec]', fontsize=20)
plt.ylim([0, 30])
plt.show()
```
